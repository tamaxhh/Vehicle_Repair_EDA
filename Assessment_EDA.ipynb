## **4. Generating Tags/Features from Free Text Fields:**
  
   To generate meaningful tags from the free text fields (`CUSTOMER_VERBATIM` and `CORRECTION_VERBATIM`), we will perform the following steps:

1. **Text Cleaning**  
        Convert text to lowercase, remove punctuation, and remove numerical digits.

2. **Tokenization**  
        Split the text into individual words.

3. **Stop Word Removal**  
        Eliminate common words (e.g., "the", "is", "and") that do not carry significant meaning.

4. **Frequency Analysis**  
        Count the occurrences of remaining words to identify the most frequent terms, which can serve as tags.

We will focus on identifying tags related to common failure conditions and components. A combined analysis of both `CUSTOMER_VERBATIM` and `CORRECTION_VERBATIM` will provide a comprehensive view of issues reported by customers and the corresponding repairs.

```python
# Import necessary libraries

import re
from nltk.corpus import stopwords
import nltk
from collections import Counter

# Download NLTK stop words (only if not already downloaded)
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    if pd.isna(text):
        return []
    text = str(text).lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation and digits
    words = text.split()  # Tokenize
    words = [word for word in words if word not in stop_words and len(word) > 2]  # Remove stop words and short words
    return words
```

```python
# Apply to both CUSTOMER_VERBATIM and CORRECTION_VERBATIM
df['CUSTOMER_VERBATIM_PROCESSED'] = df['CUSTOMER_VERBATIM'].apply(preprocess_text)
df['CORRECTION_VERBATIM_PROCESSED'] = df['CORRECTION_VERBATIM'].apply(preprocess_text)
```

```python
# Combine all processed words
all_customer_words = [word for sublist in df['CUSTOMER_VERBATIM_PROCESSED'] for word in sublist]
all_correction_words = [word for sublist in df['CORRECTION_VERBATIM_PROCESSED'] for word in sublist]

# Merge both
all_words = all_customer_words + all_correction_words

# Frequency distribution
word_freq = Counter(all_words)

# Display top 20 most frequent words
print("\nTop 20 Most Common Words (Potential Tags):\n")
for word, freq in word_freq.most_common(20):
    print(f"{word}: {freq}")
```

```python
# Common component-related and condition-related keywords
component_keywords = ['wheel', 'steering', 'module', 'engine', 'transmission', 'brake', 'light', 'sensor', 'system']
condition_keywords = ['noise', 'leak', 'vibration', 'failure', 'inop', 'check', 'message', 'replace', 'repair', 'issue']
```

```python
def generate_tags(text_list, keywords):
    tags = []
    for word in text_list:
        if word in keywords:
            tags.append(word)
    return list(set(tags))  # Return unique tags
```

```python
df['CUSTOMER_TAGS'] = df['CUSTOMER_VERBATIM_PROCESSED'].apply(lambda x: generate_tags(x, component_keywords + condition_keywords))
df['CORRECTION_TAGS'] = df['CORRECTION_VERBATIM_PROCESSED'].apply(lambda x: generate_tags(x, component_keywords + condition_keywords))
```
